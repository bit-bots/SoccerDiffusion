```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sqlite3

con = sqlite3.connect("./db.sqlite3")
```

```python
joints = ["HeadPan", "HeadTilt"]

joint_states = pd.read_sql_query(
    f"SELECT _id, {",".join(joints)} from JointStates WHERE _id >= 3000 AND _id <= 3400", con
)
joint_commands = pd.read_sql_query(
    f"SELECT _id, {",".join(joints)} from JointCommands WHERE _id >= 3000 AND _id <= 3400", con
)
```

```python
plt.plot(joint_commands["_id"], joint_commands["HeadPan"], label="HeadPanCommands")
plt.plot(joint_states["_id"], joint_states["HeadPan"], label="HeadPanStates")
plt.legend()
```

```python
plt.plot(joint_commands["_id"], joint_commands["HeadTilt"], label="HeadTiltCommands")
plt.plot(joint_states["_id"], joint_states["HeadTilt"], label="HeadTiltStates")
plt.legend()
```
